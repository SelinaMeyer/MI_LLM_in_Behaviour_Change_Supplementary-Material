```python
forum = pd.read_csv("topic_metrics/Forum_classifier_metrics_test_sets.csv", index_col="Unnamed: 0")
synthetic = pd.read_csv("topic_metrics/synthetic_classif_metrics_test_sets.csv", index_col="Unnamed: 0")
mixed_as_q = pd.read_csv("topic_metrics/mixed_as_q_classif_metrics_test_sets.csv", index_col="Unnamed: 0")
mixed_predicted = pd.read_csv("topic_metrics/mixed_predicted_classif_metrics_test_sets.csv", index_col="Unnamed: 0")
devset = pd.read_csv("topic_metrics/wandb_export_2022-03-29T13_52_54.789+02_00.csv")
devset_10_percent = pd.read_csv("topic_metrics/10_percent_wandb.csv")
```

```python
forum_2_percent = pd.read_csv("topic_metrics/forum_2_percent_classifier_metrics.csv", index_col="Unnamed: 0")
forum_4_percent = pd.read_csv("topic_metrics/forum_4_percent_classifier_metrics.csv", index_col="Unnamed: 0")
forum_6_percent = pd.read_csv("topic_metrics/forum_6_percent_classifier_metrics.csv", index_col="Unnamed: 0")
forum_8_percent = pd.read_csv("topic_metrics/forum_8_percent_classifier_metrics.csv", index_col="Unnamed: 0")
forum_10_percent = pd.read_csv("topic_metrics/forum_10_percent_classifier_metrics.csv", index_col="Unnamed: 0")
```

```python
forum_2_f1 = forum_2_percent[["Unnamed: 1", "f1"]].copy()
forum_4_f1 = forum_4_percent[["Unnamed: 1", "f1"]].copy()
forum_6_f1 = forum_6_percent[["Unnamed: 1", "f1"]].copy()
forum_8_f1 = forum_8_percent[["Unnamed: 1", "f1"]].copy()
forum_10_f1 = forum_10_percent[["Unnamed: 1", "f1"]].copy()

forum_2_f1["% of User Dataset used for Training"] = 2
forum_4_f1["% of User Dataset used for Training"] = 4
forum_6_f1["% of User Dataset used for Training"] = 6
forum_8_f1["% of User Dataset used for Training"] = 8
forum_10_f1["% of User Dataset used for Training"] = 10

forum_perc_overview = pd.concat([forum_2_f1, forum_4_f1, forum_6_f1, forum_8_f1, forum_10_f1])
forum_perc_overview.rename(columns={"Unnamed: 1": "Test Set"}, inplace=True)
forum_perc_overview["Test Set"] = forum_perc_overview["Test Set"].astype(str)
forum_perc_overview.sort_values(["% of User Dataset used for Training"], inplace=True)
forum_perc_overview["% of User Dataset used for Training"] = forum_perc_overview["% of User Dataset used for Training"].astype(str)
forum_perc_overview["Classifier"] = "User"
```

```python
mixed_2_percent = pd.read_csv("topic_metrics/mixed_predicted_2_percent_classifier_metrics.csv", index_col="Unnamed: 0")
mixed_4_percent = pd.read_csv("topic_metrics/mixed_predicted_4_percent_classifier_metrics.csv", index_col="Unnamed: 0")
mixed_6_percent = pd.read_csv("topic_metrics/mixed_predicted_6_percent_classifier_metrics.csv", index_col="Unnamed: 0")
mixed_8_percent = pd.read_csv("topic_metrics/mixed_predicted_8_percent_classifier_metrics.csv", index_col="Unnamed: 0")
mixed_10_percent = pd.read_csv("topic_metrics/mixed_predicted_10_percent_classifier_metrics.csv", index_col="Unnamed: 0")

mixed_2_f1 = mixed_2_percent[["Unnamed: 1", "f1"]].copy()
mixed_4_f1 = mixed_4_percent[["Unnamed: 1", "f1"]].copy()
mixed_6_f1 = mixed_6_percent[["Unnamed: 1", "f1"]].copy()
mixed_8_f1 = mixed_8_percent[["Unnamed: 1", "f1"]].copy()
mixed_10_f1 = mixed_10_percent[["Unnamed: 1", "f1"]].copy()

mixed_2_f1["% of User Dataset used for Training"] = 2
mixed_4_f1["% of User Dataset used for Training"] = 4
mixed_6_f1["% of User Dataset used for Training"] = 6
mixed_8_f1["% of User Dataset used for Training"] = 8
mixed_10_f1["% of User Dataset used for Training"] = 10

mixed_perc_overview = pd.concat([mixed_2_f1, mixed_4_f1, mixed_6_f1, mixed_8_f1, mixed_10_f1])
mixed_perc_overview.rename(columns={"Unnamed: 1": "Test Set"}, inplace=True)
mixed_perc_overview["Test Set"] = mixed_perc_overview["Test Set"].astype(str)
mixed_perc_overview.sort_values(["% of User Dataset used for Training"], inplace=True)
mixed_perc_overview["% of User Dataset used for Training"] = mixed_perc_overview["% of User Dataset used for Training"].astype(str)
mixed_perc_overview["Classifier"] = "Mixed Predicted"
```

```python
mixed_q_2_percent = pd.read_csv("topic_metrics/mixed_q_2_percent_classifier_metrics.csv", index_col="Unnamed: 0")
mixed_q_4_percent = pd.read_csv("topic_metrics/mixed_q_4_percent_classifier_metrics.csv", index_col="Unnamed: 0")
mixed_q_6_percent = pd.read_csv("topic_metrics/mixed_q_6_percent_classifier_metrics.csv", index_col="Unnamed: 0")
mixed_q_8_percent = pd.read_csv("topic_metrics/mixed_q_8_percent_classifier_metrics.csv", index_col="Unnamed: 0")
mixed_q_10_percent = pd.read_csv("topic_metrics/mixed_q_10_percent_classifier_metrics.csv", index_col="Unnamed: 0")

mixed_q_2_f1 = mixed_q_2_percent[["Unnamed: 1", "f1"]].copy()
mixed_q_4_f1 = mixed_q_4_percent[["Unnamed: 1", "f1"]].copy()
mixed_q_6_f1 = mixed_q_6_percent[["Unnamed: 1", "f1"]].copy()
mixed_q_8_f1 = mixed_q_8_percent[["Unnamed: 1", "f1"]].copy()
mixed_q_10_f1 = mixed_q_10_percent[["Unnamed: 1", "f1"]].copy()

mixed_q_2_f1["% of User Dataset used for Training"] = 2
mixed_q_4_f1["% of User Dataset used for Training"] = 4
mixed_q_6_f1["% of User Dataset used for Training"] = 6
mixed_q_8_f1["% of User Dataset used for Training"] = 8
mixed_q_10_f1["% of User Dataset used for Training"] = 10

mixed_q_perc_overview = pd.concat([mixed_q_2_f1, mixed_q_4_f1, mixed_q_6_f1, mixed_q_8_f1, mixed_q_10_f1])
mixed_q_perc_overview.rename(columns={"Unnamed: 1": "Test Set"}, inplace=True)
mixed_q_perc_overview["Test Set"] = mixed_q_perc_overview["Test Set"].astype(str)
mixed_q_perc_overview.sort_values(["% of User Dataset used for Training"], inplace=True)
mixed_q_perc_overview["% of User Dataset used for Training"] = mixed_q_perc_overview["% of User Dataset used for Training"].astype(str)
mixed_q_perc_overview["Classifier"] = "Mixed"

all_perc_overview = pd.concat([forum_perc_overview, mixed_perc_overview, mixed_q_perc_overview])
```

### Valence

```python
orig = pd.read_csv("valence_metrics/Sentence_test_sets_metrics.csv")
mixed_increased_balanced = pd.read_csv("valence_metrics/mixed_increased_pred_balancer.csv")
mixed_balanced = pd.read_csv("valence_metrics/mixed_pred_balancer.csv")
mixed = pd.read_csv("valence_metrics/mixed_pred.csv")
```

```python
rows = ["GLoHBCD", "Instagram Data", "Smoke Stop Forum", "Synthetic GPT3 Data", 
        "Health Coach Dialogue Corpus", "DARN-CT-based Wizard of Oz Dialogues", 
        "Optifast Mock-Chatbot"]
orig = orig[orig["Unnamed: 1"].isin(rows)].copy()
mixed_increased_balanced = mixed_increased_balanced[mixed_increased_balanced["Unnamed: 1"].isin(rows)].copy()
mixed_balanced[mixed_balanced["Unnamed: 1"].isin(rows)].copy()
mixed[mixed["Unnamed: 1"].isin(rows)].copy()
orig.head()
```

```python
mixed_increased_balanced["condition"] = "Increased Mixed Predicted Balanced"
orig["condition"] = "Original"
mixed_balanced["condition"] = "Mixed Predicted Balanced"
mixed["condition"] = "Mixed Predicted"

data = pd.concat([orig, mixed, mixed_balanced, mixed_increased_balanced])
```

```python
sns.set(rc={"figure.figsize":(15, 5)})
sns.set_palette("colorblind")
ax = sns.boxplot(x="Unnamed: 1", y="f1", hue="condition", data=data)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title("Valence")
plt.xlabel("Test Set")
plt.ylim(0.5, 1)
plt.ylabel("F1 Score")
#plt.huelabel("Test Set")
#plt.style.use("dark_background")
plt.xticks(rotation = 15)
plt.savefig("performance_enhancement_valence.png", transparent=False, bbox_inches="tight")
plt.show()
```

### Reason Type

```python
orig = pd.read_csv("reason-type-metrics/Sentence_test_sets_metrics.csv")
mixed = pd.read_csv("reason-type-metrics/mixed_pred_all_datasets.csv")
mixed_balanced = pd.read_csv("reason-type-metrics/mixed_pred_balanced_all_datasets.csv")
mixed_q = pd.read_csv("reason-type-metrics/mixed_as_q_all_datasets.csv")
```

```python
rows = ["GLoHBCD", "Instagram Data", "Smoke Stop Forum", "Synthetic GPT3 Data", 
        "Health Coach Dialogue Corpus", "DARN-CT-based Wizard of Oz Dialogues", 
        "Optifast Mock-Chatbot"]
orig = orig[orig["Unnamed: 1"].isin(rows)].copy()
mixed = mixed[mixed["Unnamed: 1"].isin(rows)].copy()
mixed_balanced = mixed_balanced[mixed_balanced["Unnamed: 1"].isin(rows)].copy()
mixed_q = mixed_q[mixed_q["Unnamed: 1"].isin(rows)].copy()
orig.head()
```
