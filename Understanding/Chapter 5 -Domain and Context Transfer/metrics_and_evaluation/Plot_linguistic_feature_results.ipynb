```python
from itertools import combinations
from scipy.stats import ttest_ind


ttests={}

for index, i in enumerate(rows):
    ttests[i] = {}
    ds = all_data[all_data["Unnamed: 1"] == i]
    grps = ds['classif'].unique()
    combs = combinations(grps, 2)

    ttests[i] = {
        f'{c1}_{c2}': ttest_ind(
            ds.loc[ds['classif'] == c1, 'f1'], 
            ds.loc[ds['classif'] == c2, 'f1']
        ) for c1, c2 in combs
    }
```

```python
ttests
```

```python
ttests_df = pd.DataFrame.from_dict({(i,j): ttests[i][j] 
                           for i in ttests.keys() 
                           for j in ttests[i].keys()},
                       orient='index')

ttests_df.head()
```

```python
ttests_df.to_csv("Significance_results.csv")
```
